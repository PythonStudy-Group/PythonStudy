```python
#12장 메인

import folium
import csv
import operator

f1 = open('LOCAL_PEOPLE_DONG_201912.csv', encoding= 'utf8')
f2 = open('dong_code.csv')
f3 = open('dong_location.csv')

peo_data = csv.reader(f1)
next(peo_data)

cod_data = csv.reader(f2)
next(cod_data)
next(cod_data)

loc_data = csv.reader(f3)
next(loc_data)

peo_data = list(peo_data)
cod_data = list(cod_data)
loc_data = list(loc_data)

#데이터들의 타입 변환. str-> int 필수!
for row in peo_data:
    for i in range(1, 32):
        if i <= 2:
            row[i] = int(row[i])
        else:
            row[i] = float(row[i])

for row in cod_data:
    row[0], row[1] = int(row[0]), int(row[1])
    
for row in loc_data:
    row[2], row[-2], row[-1] = int(row[2]), float(row[-2]), float(row[-1])
            
peo_dic = {}
top10_name = [0 for i in range(10)]
top10_lat = [0 for i in range(10)]
top10_long = [0 for i in range(10)]

for row in peo_data:
    if row[2] not in peo_dic.keys():
        peo_dic[row[2]] = row[3]
    else:
        peo_dic[row[2]] += row[3]

top10 = sorted(peo_dic.items(), key=operator.itemgetter(1), reverse=True)[:10]
top10_code = []

for t in top10:
    top10_code.append(t[0])
    
print(top10_code)
    
for i in range(10):
    for row in cod_data:
        if top10_code[i] == row[1]:
            top10_name[i] = row[4]

print(top10_name)

for i in range(10):
    for row in loc_data:
        if row[3] == top10_name[i]:
            top10_lat[i], top10_long[i] = row[-1], row[-2]
            
print(top10_lat)
print(top10_long)

map_top10 = folium.Map([top10_lat[0], top10_long[0]], zoom_start=12)

for i in range(10):
    folium.Marker([top10_lat[i], top10_long[i]], tooltip=top10_name[i], 
                 icon=folium.Icon(color='red', icon='star')).add_to(map_top10)

map_top10
```

```python
#12장 추가분석

import folium
import csv
import operator

f1 = open('LOCAL_PEOPLE_DONG_201912.csv', encoding= 'utf8')
f2 = open('dong_code.csv')
f3 = open('dong_location.csv')

peo_data = csv.reader(f1)
next(peo_data)

cod_data = csv.reader(f2)
next(cod_data)
next(cod_data)

loc_data = csv.reader(f3)
next(loc_data)

peo_data = list(peo_data)
cod_data = list(cod_data)
loc_data = list(loc_data)

#데이터들의 타입 변환. str-> int 필수!
for row in peo_data:
    for i in range(1, 32):
        if i <= 2:
            row[i] = int(row[i])
        else:
            row[i] = float(row[i])

for row in cod_data:
    row[0], row[1] = int(row[0]), int(row[1])
    
for row in loc_data:
    row[2], row[-2], row[-1] = int(row[2]), float(row[-2]), float(row[-1])
            
man_dic = {}
top10_name = [0 for i in range(10)]
top10_lat = [0 for i in range(10)]
top10_long = [0 for i in range(10)]

for row in peo_data:
    if row[2] not in peo_dic.keys():
        peo_dic[row[2]] = sum(row[4:18])
    else:
        peo_dic[row[2]] += sum(row[4:18])

top10 = sorted(peo_dic.items(), key=operator.itemgetter(1), reverse=True)[:10]
top10_code = []

for t in top10:
    top10_code.append(t[0])
    
print(top10_code)
    
for i in range(10):
    for row in cod_data:
        if top10_code[i] == row[1]:
            top10_name[i] = row[4]

print(top10_name)

for i in range(10):
    for row in loc_data:
        if row[3] == top10_name[i]:
            top10_lat[i], top10_long[i] = row[-1], row[-2]
            
print(top10_lat)
print(top10_long)

map_top10 = folium.Map([top10_lat[0], top10_long[0]], zoom_start=12)

for i in range(10):
    if i < 5:
        folium.Marker([top10_lat[i], top10_long[i]], tooltip=top10_name[i], 
                 icon=folium.Icon(color='darkblue', icon='male', prefix='fa')).add_to(map_top10)
    else:
        folium.Marker([top10_lat[i], top10_long[i]], tooltip=top10_name[i], 
                 icon=folium.Icon(color='lightblue', icon='star', prefix='fa')).add_to(map_top10)
        
        
map_top10
```
